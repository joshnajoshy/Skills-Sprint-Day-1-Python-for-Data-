```python
## plotting statistical graphics
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips") # tips name comes from the data list 
```

```python
tips # you dont need to print just write the variable name and it prints
```

```python
## plotting the data 

## scatter plot 
sns.scatterplot(x="tip", y="total_bill", data=tips) # taking x axis = tip , y axis = total_bill and the data were running
```

```python
## easily see a positive correlation between the amount tiped 
## the higher the bill total the higher the tip amount 
```

```python
## correlation does not always mean causation - means does not always mean the larger bill = corelation
```

```python
# total bill and smoker
sns.scatterplot(x="smoker", y="tip", data=tips) # taking x axis = tip , y axis = total_bill and the data were running
```

```python
# days and total bill
sns.scatterplot(x="day", y="total_bill", data=tips) # taking x axis = tip , y axis = total_bill and the data were running
```

```python
# sex and tips 
sns.scatterplot(x="size", y="tip", data=tips) # taking x axis = tip , y axis = total_bill and the data were running
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day")
```

```python
## correlation between total bill and tip type 
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size='size', palette='YlGnBu')
```

```python
# histograms
sns.histplot(tips["tip"], kde=True, bins=8) # kde is the plot line
```

```python
# if we wanna show averages or counts 
# barplot - there is a correlation but doesn't mean causation cause alot of factors could be affecting this 
sns.barplot(x='sex', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

```python
## box plot
sns.boxplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge=True)
```

```python
## join plot - a combined plot - how age and income are related
sns.jointplot(x='tip', y='total_bill', data=tips)
```

```python
## have a distribution 
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='size', y='total_bill', data=tips, kind='reg')
```

```python
## a data like this will need a regretion line to help read the data better
# pairplots - can explore the comparision of multiple different sections at the same time
```

```python
sns.pairplot(tips) # scatter plot and histaplot 
```

```python
sns.pairplot(tips, hue='smoker')
```

```python
sns.pairplot(tips, hue='smoker', kind="kde")
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.barplot(x='survived', y='pclass', data=titanic, hue='sex', palette='YlGnBu')
```

```python
sns.set_theme(style="darkgrid")
sns.countplot(x="survived", data=titanic)
```
