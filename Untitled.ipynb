```python
import pandas as pd
```

```python
df_cars = pd.read_csv("messy_cars.csv")
df_cars.head()
```

```python
# Tasks:
```

```python
# Drop or fill missing values appropriately.
df_dropped_cars = df_cars.dropna()
df_dropped_cars.head()

# solution 
# df_dropped_cars["Brand"] = df_dropped_cars["Brand"].fillna("Unknown")
# df_dropped_cars["Color"] = df_dropped_cars["Color"].fillna("Unknown")
# df_dropped_cars["Money"] = df_dropped_cars["Money"].fillna(df_dropped_cars["Money"].median())
# df_dropped_cars["Miles"] = df_dropped_cars["Miles"].fillna(df_dropped_cars["Miles"].median())
```

```python
# Filter the dataset for a meaningful subset.
filtered_cars = df_dropped_cars[(df_dropped_cars["Yr"] > 2015) & (df_dropped_cars["Money"] > 30000)] 
filtered_cars.head()
```

```python
# Sort the data by miles descending.
sort_cars = df_dropped_cars.sort_values('Miles', ascending=False)
sort_cars.head()
```

```python
# Rename columns for clarity.
df_renamed_cars = sort_cars.rename(columns={"Car_num":"Car_Number", "Mod":"Model", "Yr":"Year"})
df_renamed_cars.head()
```

```python
# Save the cleaned dataset as cars_cleaned.csv.
df_renamed_cars.to_csv('cars_cleaned.csv')
```

```python
# group by and multiple agregations 
```

```python
# group by brand and select one colum before the agregation 
# group by gets the data from a specific set of data 
```

```python
brand_group = df_cars.groupby("Brand")["Miles"]
brand_group.head()
```

```python
brand_group = df_cars.groupby("Brand")["Miles"].mean().reset_index(name="Avg_mileage")
brand_group.head()
```

```python

brand_money = df_cars.groupby("Brand")["Money"].mean().reset_index(name="Avg_Price")
brand_money.head()
```

```python
brand_money = df_cars.groupby("Brand")["Money"].mean().reset_index(name="Avg_Price")
brand_money.head()
```

```python
# multiple agregastions on the brand money, car_number
brand_stats = df_cars.groupby("Brand").agg(
    avg_price=("Money", "mean"),
    avg_millage=("Miles","median"),
    total_cars =("Car_num", "count")
).reset_index()
brand_stats.head()
```

```python
multiple_cols = df_cars.groupby(["Mod", "Color"]).agg(
    avg_price = ("Money", "mean"),
    total_cars = ("Car_num", "count")
).reset_index()
multiple_cols.head()
```

```python
## Tasks:

# Group by color and calculate the average milage
group_color = df_cars.groupby("Color")["Miles"].mean().reset_index(name="Avg_Miles")
group_color.head()
```

```python
# Groud by brand and calculate the maximum price
group_brand = df_cars.groupby("Brand")["Money"].max().reset_index(name="Max price")
group_brand.head()
```

```python
# Group by brand and color and count how many cars are in each combination
group_brand_color = df_cars.groupby(["Brand", "Color"]).agg(
    Brand_Car_Count = ("Brand", "count"),
    Color_Car_Count = ("Color", "count")
).reset_index()
group_brand_color.head()
## or 
# df_cars.groupby(["Brand", "Color"])["Mod"].count().reset_index(name="Count")
```

```python
# Group by brand and calculate both the average milage and total money spent in one .agg()
group_brand_stat = df_cars.groupby("Brand").agg(
    avg_millage=("Miles", "mean"),
    total_money_spent =("Money", "sum")
).reset_index()
group_brand_stat.head()
```
