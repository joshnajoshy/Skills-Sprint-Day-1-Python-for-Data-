```python
# Reading from a csv 
## students_with_scores.csv
df = pd.read_csv("students_with_scores.csv") # allows us to read the csv file with panda 
```

```python
df.head() # gives us first 5 in nice format and prints it out without printing
```

```python
df.tail() # gives last 5 
```

```python
# methods to explore the data 
# check the dataset shape 
print(df.shape)
# column names 
print(df.columns)
# summary of statistics
print(df.describe())
# info on the datatypes and missing values 
print(df.info()) # counts non-nulls
```

```python
# indexing / slicing - title 
# .loc - uses names to get the locations and used labels for indexing not actual positions
# .iloc - uses members to do the same things 

# single column
print(df["Name"]) # prints the name column

# multiple columns
print(df[["Name", "School"]]) # to print out multiple columns in dataframe

# row by label 
print(df.loc[0]) # 0 is getting everything in index 0 which is student 1 data  

# specific rows and column
print("specific rows and columns, name & school", df.loc[0:2, ["Name","School"]])
# getting the first 3 students names and school 

## iloc
print("Print out row by position", df.iloc[0]) # prints out the first student details 

#multiple rows

print(df.iloc[0:3]) # prints out first 3 students 

# loc and iloc picks out data differently 
```

```python
#name and score of first 3 students
print(df.loc[0:2, ["Name","Score"]])
```

```python
# all students from a specific school
filtered_students = df.loc[df['School'] == 'North High']
print(filtered_students)
```

```python
# Score of the last 2 students
print(df['Score'].iloc[-2:])
#another way
print(df.iloc[-2:, df.columns.get_loc("Score")])
```

```python
# sort by descending order
print(df.sort_values('Score', ascending=False))
```

```python
# filter students with scores above 80
print(df.loc[df['Score'] > 80])
```

```python
# Average score per school
average_scores = df.groupby('School')['Score'].mean()
print(average_scores)
```

```python
# load data 
df_messy = pd.read_csv("students_data_messy.csv")
df_messy.head()
```

```python
# drop rows where any value is missing 
df_dropped = df_messy.dropna() # gets rid of all data that has a missing value
df_dropped.head() # just prints first 5 data of the data that hasnt been dropped
# drop rows where a specific column is missing
df_dropped_col = df_messy.dropna(subset=["Age"]) # gets rid of all nan values in that sepcific column
df_dropped_col.head()
```

```python
# fill the missing values with a constant 
df_messy["Age_Filled"] = df_messy["Age"].fillna(0) # creates a new column called age filled and gets column of age and sets nan value to 0
df_messy.head(20) 
```

```python
# fill missing values with a mean or median or statistics 
df_messy["Score_Filled"] = df_messy["Score"].fillna(df_messy["Score"].mean())
# creates a new column called Score_Filled gets column score of data and fills nan value with the mean in new column
df_messy.head(20)
```

```python
# filtering & Sorting 

# filter multiple conditions use and or, age and score, attendence 
filtered_students_age = df_messy[(df_messy["Score"] > 20) & (df_messy["Age"] == 17)] 
# need to have ( ) around the conditions
filtered_students_age.head()
```

```python
# Sorting 

# sort by score ascending
sort_ascending = df_messy.sort_values('Score')
sort_ascending.head()
```

```python
# sort by score decending 
sort_decending = df_messy.sort_values('Score', ascending=False)
sort_decending.head()
```

```python
# sort by age (ascending) & score (decending)
sorted_multi = df_messy.sort_values(by=['Age', 'Score'], ascending=[True, False])
sorted_multi.head()
```

```python
# Renaming columns 

# Rename specific columns 
df_renamed = df_messy.rename(columns={"Extracurricular":"Club"})
df_renamed.head()

# Rename all columns in one go
df_messy.head()
df_messy.columns = ["Name", "School_Name", "Grade_Level", "Test_Score", "Age_Year","Attendance_Rate", "Club", "Email_Address", "Age_Extra","Score_Mean"]
df_messy.head()
```
