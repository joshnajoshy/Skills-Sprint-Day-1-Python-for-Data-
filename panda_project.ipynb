```python
import pandas as pd
```

```python
df = pd.read_csv("sales_with_nans (3).csv")
df.head()
```

```python
df["Group"] = df["Group"].fillna("Unknown")
df["Customer_Segment"] = df["Customer_Segment"].fillna("Unknown")
df["Sales_Before"] = df["Sales_Before"].fillna(df["Sales_Before"].median())
df["Sales_After"] = df["Sales_After"].fillna(df["Sales_After"].median())
df["Customer_Satisfaction_Before"] = df["Customer_Satisfaction_Before"].fillna(df["Customer_Satisfaction_Before"].median())
df["Customer_Satisfaction_After"] = df["Customer_Satisfaction_After"].fillna(df["Customer_Satisfaction_After"].median())
df["Purchase_Made"] = df["Purchase_Made"].fillna("Unknown")
df.head()
```

```python
# Total sales before & after
total_sales_before = df['Sales_Before'].sum()
total_sales_after = df['Sales_After'].sum()
print(total_sales_before)
print(total_sales_after)
```

```python
# Average sales before & after
avarage_sales_before = df['Sales_Before'].mean()
avarage_sales_after = df['Sales_After'].mean()
print(avarage_sales_before)
print(avarage_sales_after)
```

```python
# Average satisfaction before & after
avarage_sales_sat_before = df['Sales_Before'].mean()
avarage_sales_sat_after = df['Sales_After'].mean()
print(avarage_sales_sat_before)
print(avarage_sales_sat_after)
```

```python
# Minimum and maximum sales values
max_sales_before = df['Sales_Before'].max()
min_sales_before = df['Sales_Before'].min()

max_sales_after = df['Sales_After'].max()
min_sales_after = df['Sales_After'].min()

print(max_sales_before, min_sales_before)
print(max_sales_after, min_sales_after)
```

```python
# Calculate the change in average sales and change in satisfaction from before to after
change_in_average_sats = avarage_sales_sat_after - avarage_sales_sat_before
print(change_in_average_sats)
change_in_average_sales = avarage_sales_after - avarage_sales_before
print(change_in_average_sales)
```

```python
# Get all rows for a specific customer_segment e.g high value
filter_specific_seg = df.loc[df['Customer_Segment'] == 'High Value']
filter_specific_seg.head()
```

```python
# Get all customers with sales above the average
filter_specific_sales_before = df.loc[df['Sales_Before'] > avarage_sales_before]
filter_specific_sales_after = df.loc[df['Sales_Before'] > avarage_sales_after]
filter_specific_sales_before.head()
filter_specific_sales_after.head()
```

```python
# Filter by multiple conditions (e.g., satisfaction â‰¥ 4 and sales > 100)
filtered_multiple_before = df[(df["Customer_Satisfaction_Before"] >= 4) & (df["Sales_Before"] > 100)]
filtered_multiple_before.head()
```

```python
# avg sales before / after customer segment
average_sales = df.groupby("Customer_Segment")["Sales_Before"].mean().reset_index(name="Avg_sales_before")
average_sales.head()
```

```python
# Satisfaction improvement by Group (Control vs Treatment)
group_sats = df.groupby(["Group"]).agg(
    total_satisfaction_after = ("Customer_Satisfaction_After", "sum"),
     total_satisfaction_before = ("Customer_Satisfaction_Before", "sum")
).reset_index()
group_sats["satisfaction_improvement"] = (group_sats["total_satisfaction_after"] - group_sats["total_satisfaction_before"])
group_sats.head()
```

```python
# Find which Customer_Segment had the biggest sales increase.
group_customer = df.groupby(["Customer_Segment"]).agg(
    total_sales_after = ("Sales_After", "sum"),
     total_sales_before = ("Sales_Before", "sum")
).reset_index()
group_customer["sales_increase"] = (group_customer["total_sales_after"] - group_customer["total_sales_before"])
group_customer.head()
sales_increase_max = group_customer["sales_increase"].max()
print(sales_increase_max)
print(group_customer.loc[group_customer['sales_increase'] == sales_increase_max])
```
