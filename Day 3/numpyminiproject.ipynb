```python
high_column = df['high'].to_numpy()
print(high_column)
```

```python
low_column = df['low'].to_numpy()
print(low_column)
```

```python
difference_in_days = high_column - low_column
```

```python
smallest_increase = np.min(difference_in_days)
biggest_increase = np.max(difference_in_days)
print(smallest_increase)
print(biggest_increase)

index_of_biggest_increase = np.argmax(difference_in_days)
index_of_smallest_increase = np.argmin(difference_in_days)
print(index_of_biggest_increase)
print(index_of_smallest_increase)
```

```python
print("date of biggest increase", date[index_of_biggest_increase])
```

```python
print("date of smallest increase", date[index_of_smallest_increase])
```
